Here 0= Normal Transaction

```python
# Here splitting the Dataset into two parts 
normal=cred_card[cred_card.Class==0]
fraud=cred_card[cred_card.Class==1]
```

```python
# sampling the Dataset.
normal_cred_card=normal.sample(n=492)
```

```python
normal_cred_card.head() 
```

```python
normal_cred_card.Class.value_counts()
```

```python
#Contact the Data 
new_credit_data=pd.concat([normal_cred_card,fraud],axis=0)
```

```python
# Independent Variable and Dependent variable
x=new_credit_data.drop("Class",axis=1)
y=new_credit_data['Class']
```

```python
# apply Logistic Regression Method
lr=LogisticRegression()
```

```python
# Assigning the value to variables.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
# Finding the shape 
print(x.shape,x_train.shape,x_test.shape)
```

```python
# Training the Data
train_data=lr.fit(x_train,y_train)
```

```python
# predicting Data
predict_data=lr.predict(x_train)
```

```python
# accuracy of train data
accuracy_train_data=accuracy_score(predict_data,y_train)
print(accuracy_train_data)
```

```python
# Predicting on Test data
predict_data_test=lr.predict(x_test)
```

```python
#accuracy on Test data
accuracy_test_data=accuracy_score(predict_data_test,y_test)
accuracy_test_data
```

As here we conclude that the accuracy of train data is similar to accuracy of test data.
